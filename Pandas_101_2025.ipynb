## HW 4 (group) ยอดรวมทั้ง 5 เดือน เมืองไหนมีจำนวนผู้ถูกเรียกค้นที่เป็นคน Asia น้อยที่สุด และ มากที่สุด


````
from google.colab import drive
import pandas as pd
import os
from collections import defaultdict

# Mount Google Drive
drive.mount('/content/drive')

# อ่านข้อมูลทั้งหมดจากโฟลเดอร์ย่อยใน base_path
def read_all_data(base_path):
    all_data = [] #สร้างlistว่างเพื่อเก็บ Dataframe ของ CSV ที่จะอ่านเข้ามา
    for month in os.listdir(base_path): #วนลูปอ่านชื่อโฟลเดอร์ย่อยทั้งหมดที่อยู่ใน base_path
        month_path = os.path.join(base_path, month)
        if not os.path.isdir(month_path):
            continue
        for file in os.listdir(month_path):
            if 'stop-and-search' not in file: #ตรวจสอบว่าชื่อไฟล์มีคำว่า stop-and-search
                continue
            stopandsearchpath = os.path.join(month_path, file)
            try:
                df = pd.read_csv(stopandsearchpath)
                # ตัดชื่อไฟล์เพื่อเอาแค่ชื่อเมือง เช่น "avon-and-somerset"
                city = file.replace('-stop-and-search.csv', '').split('-', 2)[-1].strip()
                df['Police Force'] = city #ใส่ชื่อเมืองที่ได้จากการตัดชื่อไฟล์เมื่อกี้เข้าไปทุกแถว
                all_data.append(df) #เพิ่ม Dataframe ที่อ่านแล้วเข้าไปใน all_data
            except Exception as e:
                print(f"Error reading {stopandsearchpath}: {e}")
    return all_data

# นับจำนวนคนเชื้อสายเอเชีย
def count_asian_people(all_dfs):
    asian_counts = defaultdict(int) #สร้าง defaultdict ที่มีค่าเริ่มต้นเป็น 0
    for df in all_dfs:
        asian_df = df[df['Self-defined ethnicity'].str.contains('Asian', na=False)] #กรองข้อมูลที่มีคำว่า asian
        if not asian_df.empty:
            city = asian_df['Police Force'].iloc[0]
            asian_counts[city] += len(asian_df)
    return dict(asian_counts)

# === Main execution ===
base_path = '/content/drive/MyDrive/Python&DataViz2025'

# โหลดข้อมูล
dataframes = read_all_data(base_path)

# คำนวณจำนวนคนเชื้อสายเอเชีย
counts = count_asian_people(dataframes)

# แสดงผล
if counts:
    most_city = max(counts, key=counts.get)
    fewest_city = min(counts, key=counts.get)

    print(f"เมืองที่มีจำนวนผู้ถูกเรียกค้นที่เป็นคน Asia น้อยที่สุด คือเมือง {fewest_city} จำนวน {counts[fewest_city]} คน")
    print(f"เมืองที่มีจำนวนผู้ถูกเรียกค้นที่เป็นคน Asia มากที่สุด คือเมือง {most_city} จำนวน {counts[most_city]} คน")
else:
    print("ไม่พบข้อมูลที่เกี่ยวข้องกับผู้มีเชื้อสายเอเชีย")
````

## HW 5 (เดี่ยว) ***แก้จาก โค้ดเดิม*** ยอดรวมทุกเมือง เดือนไหนมีจำนวนผู้ถูกเรียกค้นที่เป็นคน Asia น้อยที่สุด และ มากที่สุด

````
from google.colab import drive
import pandas as pd
import os
from collections import defaultdict

# Mount Google Drive
drive.mount('/content/drive')

# อ่านข้อมูลทั้งหมดจากโฟลเดอร์ย่อยใน base_path
def read_all_data(base_path):
    all_data = []
    for month in os.listdir(base_path):
        month_path = os.path.join(base_path, month)
        if not os.path.isdir(month_path):
            continue
        for file in os.listdir(month_path):
            if 'stop-and-search' not in file:
                continue
            stopandsearchpath = os.path.join(month_path, file)
            try:
                df = pd.read_csv(stopandsearchpath)
                df['month'] = month # เพิ่มคอลัมน์ month
                all_data.append(df)
            except Exception as e:
                print(f"Error reading {stopandsearchpath}: {e}")
    return all_data

# นับจำนวนคนเชื้อสายเอเชียในแต่ละเดือน
def count_asian_people_by_month(all_dfs):
    asian_counts = defaultdict(int)
    for df in all_dfs:
        asian_df = df[df['Self-defined ethnicity'].str.contains('Asian', na=False)]
        if not asian_df.empty:
            month = asian_df['month'].iloc[0] #เปลี่ยนจาก city เป็น month
            asian_counts[month] += len(asian_df)
    return dict(asian_counts)

# === Main execution ===
base_path = '/content/drive/MyDrive/Python&DataViz2025'

# โหลดข้อมูล
dataframes = read_all_data(base_path)

# คำนวณจำนวนคนเชื้อสายเอเชียแยกตามเดือน
counts = count_asian_people_by_month(dataframes)

# แสดงผล
if counts:
    most_month = max(counts, key=counts.get)
    fewest_month = min(counts, key=counts.get)

    print(f"เดือนที่มีจำนวนผู้ถูกเรียกค้นที่เป็นคน Asia น้อยที่สุด คือเดือน {fewest_month} จำนวน {counts[fewest_month]} คน")
    print(f"เดือนที่มีจำนวนผู้ถูกเรียกค้นที่เป็นคน Asia มากที่สุด คือเดือน {most_month} จำนวน {counts[most_month]} คน")
else:
    print("ไม่พบข้อมูลที่เกี่ยวข้องกับผู้มีเชื้อสายเอเชีย")
````

````
# ตรวจสอบและแสดงผลลัพธ์
print("📊 จำนวนไฟล์ข้อมูล stop-and-search ในแต่ละเดือน:\n")

# Define the 'months' variable by listing directories in base_path
months = [d for d in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, d))]
months.sort() # Optional: sort the months for consistent output

for month in months:
    month_path = os.path.join(base_path, month)

    if os.path.exists(month_path):
        files = [f for f in os.listdir(month_path) if f.endswith('-stop-and-search.csv')]
        city_names = [f.replace(f"{month}-", '').replace('-stop-and-search.csv', '') for f in files]
        print(f"{month}: {len(files)} เมือง")
        print(f"เมืองที่มีข้อมูล: {', '.join(sorted(city_names))}\n")
    else:
        print(f"{month}: ❌ ไม่พบโฟลเดอร์เดือนนี้\n")
````

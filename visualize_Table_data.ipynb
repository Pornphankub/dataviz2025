```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
display(iris_df.head())
```

```python
iris_df = iris_df.drop('target', axis=1)
```

```python
iris_df
```

#Heatmap

```python
from matplotlib import pyplot as plt
import matplotlib
```

```python
matplotlib.rcParams['figure.figsize'] = (15,30)
plt.imshow(iris_df.iloc[:,:],interpolation='nearest')
```

# pandas : parallel coordinate

```python
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
display(iris_df.head())
```

```python
matplotlib.rcParams['figure.figsize'] = (7,7)
pd.plotting.parallel_coordinates(iris_df, 'target')
```

# pandas : scatter matrix

```python
matplotlib.rcParams['figure.figsize'] = (9,9)

pd.plotting.scatter_matrix(iris_df);
```

# Color

```python
matplotlib.rcParams['figure.figsize'] = (7,7)
pd.plotting.parallel_coordinates(iris_df, 'target',color=['c','m','y']);
```

```python
matplotlib.rcParams['figure.figsize'] = (7,7)
pd.plotting.parallel_coordinates(iris_df, 'target',color= [[1,0,0,],[0.5,0.1,1],[0,1,1]]);
```

```python
matplotlib.rcParams['figure.figsize'] = (7,7)
pd.plotting.parallel_coordinates(iris_df, 'target',color= ['#013168','#A22D71','#B9A023']); #html color code
```

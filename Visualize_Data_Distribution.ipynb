````
stopandsearch_df_dropna = stopandsearch_df.dropna(subset=['Age range','Gender'])
````

````
#replace values in gender : male with 1 female with 0
stopandsearch_df_dropna['Gender'] = stopandsearch_df_dropna['Gender'].replace({'Male':35,'Female':1})
#replace values in Age range : under10 with 0 10-17 with 1 18-24 with2 25-34 eith3 over34 with4
stopandsearch_df_dropna['Age range'] = stopandsearch_df_dropna['Age range'].replace({'under 10':10,'10-17':14,'18-24':21,'25-34':30,'over 34':34})
````

สร้าง list ของตัวเลข random number ที่เป็น normal dis

````
import numpy as np
list_of_rn = np.random.normal(loc=1, scale=0.1, size=len(stopandsearch_df_dropna))
````

````
plt.hist(list_of_rn)
````

````
hist_out
````

````
hist_out=plt.hist(list_of_rn,bins=20)
````

````
stopandsearch_df_dropna['Gender'] = stopandsearch_df_dropna['Gender'] * list_of_rn
````

````
plt.subplot(3,3,1)
plt.boxplot(stopandsearch_df['Latitude'].dropna(),whis=1,showmeans=True,meanline=True,vert=False)
plt.title('boxplot')
plt.subplot(3,3,5)
plt.scatter(stopandsearch_df['Longitude'],stopandsearch_df['Latitude'],alpha=0.5)
plt.title('scatter')
plt.subplot(3,3,9)
plt.hist(stopandsearch_df['Longitude'].dropna())
plt.title('hist')
plt.tight_layout()
````

````
plt.figure(figsize=(10, 8))

# Boxplot in the top-left corner (1,1)
plt.subplot(2, 2, 1)
plt.boxplot(stopandsearch_df['Latitude'].dropna(), whis=1, showmeans=True, meanline=True, vert=False)
plt.title('Boxplot of Latitude')

# Scatter plot in the bottom-right corner (2,2)
plt.subplot(2, 2, 4)
# For a scatter plot of a single variable, we can plot against the index or a series of ones.
plt.scatter(stopandsearch_df.index, stopandsearch_df['Latitude'].dropna(), alpha=0.5)
plt.title('Scatter Plot of Latitude')

# Histogram in the bottom-left corner (2,1)
plt.subplot(2, 2, 3)
plt.hist(stopandsearch_df['Latitude'].dropna(), bins=20)
plt.title('Histogram of Latitude')

plt.tight_layout()
plt.show()
````
